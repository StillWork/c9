```python
tips = sns.load_dataset("tips")
sns.barplot(x = "day", y = "total_bill", data = tips)
```

## 파라미터 설정에는 두가지 종류가 있다.
- Styling
- Scaling

```python
sns.set_style("whitegrid")
sns.boxplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.set_style("ticks")
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.set_style("white") # 디폴트 설정
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.set_style("dark")
sns.boxplot(x="day", y="total_bill", data=tips)
```

```python
sns.set_style("ticks")
sns.boxplot(x="day", y="total_bill", data=tips)
sns.despine()
# 우측, 상위 라인을 없앤다
```

```python
sns.set_style("ticks")
sns.boxplot(x="day", y="total_bill", data=tips)
sns.despine(left=True)
```

하나의 그림에서 두가지 배경을 설정하기 위해서 axes_style()을 사용한다.

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
# 그림 1
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
    
# 그림 2
plt.subplot(212)
sinplot(-1)
```

`paper`, `notebook(디폴트)`, `talk`, `poster`

```python
# 환경 재설정
sns.set()
```

```python
sns.set_style("whitegrid")
sns.set_context("paper")

sns.boxplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.set_context("poster")
sns.boxplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.set_context("notebook")
sns.boxplot(x = "day", y = "total_bill", data = tips)
```

```python
sns.set_context("talk")
sns.boxplot(x = "day", y = "total_bill", data = tips)
```

## 그림의 크기가 달라지지 않는 것은 쥬피터에서 조정하기 때문이다.
- rc 으로도 시각적 효과를 바꿀 수 있다.

```python
sns.set(style='whitegrid', rc={"grid.linewidth": 1.5})
sns.set_context("poster", font_scale=2.5, rc={"lines.linewidth": 5.0})
sinplot()
```
